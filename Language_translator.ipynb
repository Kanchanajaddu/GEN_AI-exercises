```python
import openai
```

```python
def translate_text(text, target_language='English', source_language='auto'):
    try:
        if source_language == 'auto':
            prompt = f"Translate the following text into {target_language}:\n\n'{text}'"
        else:
            prompt = f"Translate the following text from {source_language} into {target_language}:\n\n'{text}'"

        response = openai.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}]
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"An error occurred during translation: {e}"
```

### Demo of the Translator

```python
text_to_translate = "Hola, ¿cómo estás?"
target_lang = "French"

translated_text = translate_text(text_to_translate, target_lang)
print(f"Original ({text_to_translate}): {text_to_translate}")
print(f"Translated ({target_lang}): {translated_text}")

print("\n---")

text_to_translate_2 = "Je m'appelle John."
target_lang_2 = "Spanish"

translated_text_2 = translate_text(text_to_translate_2, target_lang_2)
print(f"Original ({text_to_translate_2}): {text_to_translate_2}")
print(f"Translated ({target_lang_2}): {translated_text_2}")
```
